```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Les CSV-fila
def les_csv(filnavn):
    return pd.read_csv(filnavn)

# Lag stolpediagram for filosofar
def lag_filosof_stolpediagram(data):
    filosofar = data['Navn']
    antall = Counter(filosofar)
    
    plt.figure(figsize=(10, 6))
    plt.bar(antall.keys(), antall.values(), color='skyblue')
    plt.xlabel('Filosofar')
    plt.ylabel('Antall')
    plt.title('Filosofar - Antall gongar nemnt')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Lag stolpediagram for tema
def lag_tema_stolpediagram(data):
    tema = data['Tema'].str.split(', ').sum()
    antall = Counter(tema)
    
    plt.figure(figsize=(10, 6))
    plt.bar(antall.keys(), antall.values(), color='orange')
    plt.xlabel('Tema')
    plt.ylabel('Antall')
    plt.title('Tema - Antall gongar nemnt')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

# Lag oppsummeringstabell
def lag_oppdateringstabell(data):
    oppsummering = data.groupby('Navn')['Tema'].apply(lambda x: ', '.join(x)).reset_index()
    oppsummering['Antall'] = data['Navn'].value_counts().values
    display(oppsummering)
    return oppsummering

# Kjøring
filnavn = "filosofer.csv"  # Set filnamnet her
data = les_csv(filnavn)

# Lag diagram og tabell
lag_filosof_stolpediagram(data)
lag_tema_stolpediagram(data)
oppsummeringstabell = lag_oppdateringstabell(data)

# Eksporter oppsummering til CSV om ønska
oppsummeringstabell.to_csv("oppsummering.csv", index=False)
```
